```python
class NoArvore:
  valor = None
  esq = None
  dir = None

def Arvore_add(no, val):
  if val < no.valor:
    if no.esq:
      Arvore_add(no.esq, val)

    else:
      no.esq = NoArvore()
      no.esq.valor = val

  else:
    if no.dir:
      Arvore_add(no.dir, val)

    else:
      no.dir = NoArvore()
      no.dir.valor = val



def arvore_print(no, nivel=1):
  if no:
    arvore_print(no.dir, nivel+1)
    print(' ' * 4 * nivel + '->' + str(no.valor))
    arvore_print(no.esq, nivel+1)







def arvore_ordem(no):
  if no:
    arvore_ordem(no.esq)
    print(no.valor)
    arvore_ordem(no.dir)

def arvore_pre_ordem(no):
  if no:
    arvore_pre_ordem(no.dir)
    print('_', no.valor)
    arvore_pre_ordem(no.esq)



def arvore_pos_ordem(no):
  if no:
    arvore_pos_ordem(no.esq)
    arvore_pos_ordem(no.dir)
    print('-', no.valor)


def getfilhos(nos):
  filhos = []
  for no in nos:
    if no.esq:
      filhos.append(no.esq)
    if no.dir:
      filhos.append(no.dir)

  return filhos


def print_filhos(filhos):
  for no in filhos:
    print(no.valor, end=' ')
  print()



def arvore_print_largura(no):
  if no:
    print(no.valor)
    filhos = getfilhos([no])
    while filhos:
      print_filhos(filhos)
      filhos = getfilhos(filhos)


raiz = NoArvore()
raiz.valor = 14
Arvore_add(raiz,12)
Arvore_add(raiz,30)
Arvore_add(raiz,0)
Arvore_add(raiz,5)
Arvore_add(raiz,9)
Arvore_add(raiz,18)
Arvore_add(raiz,12)
Arvore_add(raiz,22)
Arvore_add(raiz,21)
Arvore_add(raiz,20)
arvore_print(raiz)
print('\n \n \n')
arvore_print_largura(raiz)
```
