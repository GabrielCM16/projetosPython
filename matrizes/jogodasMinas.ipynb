```python

#numMinas = int(input("quantas minas tera o jogo "))
print()

def criaTabuleiro(tam=5):

  matrizes = dict()

  for i in range(tam):
    matriz = 'matriz' + str(i+1)
    matrizes[matriz] = tam * ['X']

  for i in range(tam):
    matriz = 'matriz' + str(i+1)
    print(matrizes[matriz])

  print('\n')
  return matrizes




def modificaTabuleiro(matrizes, x, y, muda='O'):

  matriz_requisitada = 'matriz' + str(x)

  matrizes[matriz_requisitada][int(y) - 1] = muda

  for i in range(len(matrizes)):
    matriz = 'matriz' + str(i+1)
    print(matrizes[matriz])






matrizes = criaTabuleiro()
modificaTabuleiro(matrizes, 5, 5)
```

```python
import random
def sorteiaPosMina(tam_matrizes, N_sort):

  numeros_sort = list()

  while len(numeros_sort) < N_sort:
    x = random.randint(0, tam_matrizes)
    y = random.randint(0, tam_matrizes)

    lista_temp = [x, y]

    if lista_temp in numeros_sort:
      continue
    else: numeros_sort.append(lista_temp)

  return numeros_sort




```

```python
tamJogo = int(input("qual sera o tamanho do tabuleiro (matriz menor que 9x9) "))
numMinas = int(input("quantas minas tera o jogo "))
print()

numeros_minas = sorteiaPosMina(tamJogo, numMinas)
matrizes = criaTabuleiro(tamJogo)
vidas = 3
while vidas > 0:

  print()
  print('abrir qual posicao da matriz (linha X coluna)')
  print('exemplo: (2 3)')
  x, y = input("").split()
  lista_temp = [int(x), int(y)]
  print()

  if lista_temp in numeros_minas:
    print()
    vidas -= 1
    print(f"errado, vidas: {vidas}")
    print()
    modificaTabuleiro(matrizes, int(x), int(y), 'W')

  else:

    modificaTabuleiro(matrizes, int(x), int(y))
```
