```python
#Cadastro e Venda de Carros
import json


import os
local = os.path.abspath(os.getcwd())
local = local+'/arquivos.json'
if os.path.isfile(local):
    print('O caminho {} existe'.format(local))
else:
    exemplo = {
    "Carros": {
        "1": {
            "Modelo": "Honda Civic",
            "Ano": "2000",
            "Valor": "10000",
            "condicao": "Vendido"
        }
    },
    "Vendedores": {
        "1": {
            "Nome": "Jonas",
            "Data_Nasc": "28/09/1987",
            "Numero_tel": "(45) 99999-9999",
            "CPF": "000.000.000-00"
        }
    },
    "Clientes": {
        "1": {
            "Nome": "Francisco",
            "Data_Nasc": "10/05/1979",
            "Numero_tel": "(45) 88888-8888"
        }
    },
    "Vendas": {
        "1": {
            "Venda_por": "1",
            "Comprado_por": "1",
            "carro": "1",
            "Valor": "9500"
        }

        }
    }


    json_file = open('arquivos.json', mode='w')
    json.dump(exemplo, json_file, indent=4)
    json_file.close()



with open('arquivos.json', 'r') as file:
  arquivos = json.load(file)

while True:


  def CadastrarCarros():
    id = list(arquivos['Carros'].keys())
    for i in range(len(id)):
        id[i] = int(id[i])

    id_novo = max(id) + 1

    modelocarro = input('Qual o modelo do Carro: ')
    anocarro = input('Qual o ano de fabricação do carro: ')
    valor = input('Qual o valor do carro: ')
    arquivos['Carros'][str(id_novo)] = {
            "Modelo" : modelocarro,
            "Ano" : anocarro,
            "Valor" : valor,
            "condicao": "Disponivel"
        }

    with open('arquivos.json', 'w') as file:
      json.dump(arquivos, file, indent=4)

    print('\n')



  def CadastrarCliente():
    id = list(arquivos['Clientes'].keys())
    for i in range(len(id)):
        id[i] = int(id[i])

    id_novo = max(id) + 1

    nomecliente = input('Nome cliente: ')
    datanasc = input('Data de nascimento cliente: ')
    numerotel = input('Numero de telefone: ')
    print('\n')

    arquivos["Clientes"][str(id_novo)] = {
            "Nome" : nomecliente,
            "Data_Nasc" : datanasc,
            "Numero_tel" : numerotel
    }

    with open('arquivos.json', 'w') as file:
      json.dump(arquivos, file, indent=4)


  def CadastrarVendedor():
    nomevendedor = input('Nome vendedor: ')
    datanasc = input('Data de nascimento: ')
    numerotel = input('Numero de telefone: ')
    cpf = input('CPF: ')
    print('\n')

    id = list(arquivos['Vendedores'].keys())
    for i in range(len(id)):
        id[i] = int(id[i])

    id_novo = max(id) + 1

    arquivos["Vendedores"][str(id_novo)] = {
            "Nome" : nomevendedor,
            "Data_Nasc" : datanasc,
            "Numero_tel" : numerotel,
            "CPF" : cpf
    }

    with open('arquivos.json', 'w') as file:
      json.dump(arquivos, file, indent=4)





  def Vender():
      print('Vendedores cadastrados: ')
      listarvendedor = list()
      for i in range(len(arquivos['Vendedores'])):
              listarvendedor.append(arquivos['Vendedores'][str(i+1)]["Nome"])

      for i in range(len(listarvendedor)):
          print('Vendedor: id =',i+1,listarvendedor[i])

      print('\n')
      ven = int(input('Quem realizou a venda: (id) '))
      print('\n')

      print('Carros cadastrados disponiveis: ')
      listarcarro = list()
      for i in range(len(arquivos['Carros'])):
            if arquivos['Carros'][str(i+1)]["condicao"] == "Disponivel":
                listarcarro.append(arquivos['Carros'][str(i+1)])

      for i in range(len(listarcarro)):
          print('Carro: id =',i+1,listarcarro[i])
      print('\n')
      carr = int(input('Qual carro foi vendido: (id) '))
      print('\n')

      listarcli = list()
      for i in range(len(arquivos['Clientes'])):
              listarcli.append(arquivos['Clientes'][str(i+1)])

      for i in range(len(listarcli)):
          print('Cliente: id =',i+1,listarcli[i])

      print('\n')
      clien = int(input('Qual cliente comprou: (id) '))


      id = list(arquivos['Vendas'].keys())
      for i in range(len(id)):
          id[i] = int(id[i])

      id_novo = max(id) + 1

      arquivos["Vendas"][str(id_novo)] = {
            "Venda_por": str(ven),
            "Comprado_por": str(clien),
            "carro": str(carr),
            "Valor": input('Qual o valor de negocio fechado: ')
      }

      arquivos["Carros"][str(carr)]["condicao"] = "Vendido"


      with open('arquivos.json', 'w') as file:
        json.dump(arquivos, file, indent=4)


  def VendaVende():
      print('Vendedores cadastrados: ')

      for i in range(len(arquivos["Vendedores"])):
          print('Vendedor: id =',i+1,arquivos["Vendedores"][str(i+1)]["Nome"])
      print('\n')
      x = input('Qual vendedor deseja listar: id(): ')
      print('\n')
      numv = 0
      for i in range(len(arquivos["Vendas"])):
             if arquivos["Vendas"][str(i+1)]["Venda_por"] == x:
                  numv += 1
                  print(arquivos["Vendas"][str(i+1)])
      nome = arquivos["Vendedores"][x]["Nome"]
      print('\n')
      print(f"O vendedor id='{x}' ({nome}) realizou {numv} venda(s) ao total. ")
      print('\n')



  def listar():
    print('Oque deseja listar \n')
    print('[1] Carros não vendidos')
    print('[2] Carros já vendidos')
    print('[3] Vendas por vendedor')
    print('[4] Carros por valor ')
    fazer4 = int(input())

    listarcarrosN = list()
    listarcarrosV = list()

    for i in range(len(arquivos['Carros'])):
            if arquivos['Carros'][str(i+1)]["condicao"] == "Disponivel":
                listarcarrosN.append(arquivos['Carros'][str(i+1)])
            else: listarcarrosV.append(arquivos['Carros'][str(i+1)])

    if fazer4 == 1:
        print('\n')
        print('Carros disponiveis: ')
        for i in range(len(listarcarrosN)):
                print(listarcarrosN[i])


    if fazer4 == 2:
            print('\n')
            print('Carros que já foram vendidos: ')
            for i in range(len(listarcarrosV)):
                print(listarcarrosV[i])

    if fazer4 == 3:
        VendaVende()
    listarcarrosval = list()

    if fazer4 == 4:
        print('\n')
        print('Carros por valor: ')
        val = int(input('Digite um valor: '))
        for i in range(len(arquivos['Carros'])):
            valorc = arquivos['Carros'][str(i+1)]["Valor"]
            if int(valorc) <= val:
                listarcarrosval.append(arquivos['Carros'][str(i+1)])
        print('\n')
        for i in range(len(listarcarrosval)):
            print(listarcarrosval[i])



  def Cadastro():
    print('Oque deseja cadastrar: \n')
    print(' [1] Carros')
    print(' [2] Clientes')
    print(' [3] Vendedor')
    cadas = int(input())

    if cadas == 1:
      CadastrarCarros()
    if cadas == 2:
      CadastrarCliente()
    if cadas == 3:
      CadastrarVendedor()

  print('Oque deseja fazer: \n')
  print(' [1] Cadastrar')
  print(' [2] Vender')
  print(' [3] Listar')
  print(' [4] Sair do sistema')
  fazer = int(input())

  if fazer == 4:
    break
  elif fazer == 1:
    Cadastro()
  elif fazer == 3:
    listar()
  elif fazer == 2:
      Vender()


```
